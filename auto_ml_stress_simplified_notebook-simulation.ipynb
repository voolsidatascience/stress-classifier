```python
def fix_stress_labels(stress):
    if 0 <= stress < 0.2:
        target = 0
    elif 0.2 <= stress < 0.4:
        target = 1
    elif 0.4 <= stress < 0.6:
        target = 2
    elif 0.6 <= stress < 0.8:
        target = 3
    else:
        target = 4
    return target


dataframe_hrv['target'] = dataframe_hrv['stress'].apply(fix_stress_labels)
dataframe_hrv.head()
```

```python
def missing_values(df):
    #df = df.reset_index()
    df = df.replace([np.inf, -np.inf], np.nan)
    #df[~np.isfinite(df)] = np.nan
    #df['HR'].fillna((df['HR'].mean()), inplace=True)
    
    df = df.apply(lambda x: x.fillna(x.mean()),axis=0)
    df['HR'] = signal.medfilt(df['HR'], 13) 
    return df

dataframe = missing_values(dataframe_hrv)
```
